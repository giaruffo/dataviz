```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [0,1,2,3,4,5]
y = [0,1,4,9,16,25]
```

```python
plt.plot(x,y)
```

```python
plt.show()
```

```python
plt.plot(x,y)
plt.title("first plot")
plt.xlabel("X")
plt.ylabel("Y = X^2")
plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,"go")
plt.title("first plot")
plt.xlabel("X")
plt.ylabel("Y = X^2")
plt.show()
```

```python
x1 = np.arange(0,6)
y1 = x1**3
```

```python
plt.plot(x,y,"go", x1, y1, "r^")
plt.title("first plot")
plt.xlabel("X")
plt.ylabel("Y = X^2")
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,"go")
plt.title("1st sub plot")
plt.subplot(1,2,2)
plt.plot(x1, y1, "r^")
plt.title("2nd sub plot")
plt.suptitle("My sub plots")
plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(x,y,"go")
plt.title("1st sub plot")
plt.subplot(2,1,2)
plt.plot(x1, y1, "r^")
plt.title("2nd sub plot")
plt.suptitle("My sub plots")
plt.show()
```

```python
x = np.arange(1,10)

fig,ax=plt.subplots(nrows=2,ncols=2,figsize=(10,10))

y2 = np.log(x)
ax[0,0].plot(x,y2,"b-")
ax[0,0].set_title("log")

y3 = np.log10(x)
ax[0,1].fill_between( x, y3, color="skyblue", alpha=0.2)
ax[0,1].plot(x, y3, color="Slateblue", alpha=0.6)
ax[0,1].set_title("log10")
ax[0,1].set_ylim(ax[0,0].get_ylim())

y = x**2
ax[1,0].plot(x,y,"go")
ax[1,0].set_title("Squares")

y1 = x**3
ax[1,1].plot(x, y1, "r^")
ax[1,1].set_title("Cubes")

ax[1,0].set_ylim(ax[1,1].get_ylim())

plt.suptitle("My sub plots")

plt.show()
```

```python
k = np.arange(1,10**6)
f = pow(k,-2.5)
plt.plot(k,f,"go")
plt.title("power law")
plt.xscale("log")
plt.yscale("log")
plt.xlabel("k")
plt.ylabel("f(k) = k^-2.5")
plt.show()
```
