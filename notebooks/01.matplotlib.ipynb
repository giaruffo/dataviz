## Plotting a line chart

Let's start with plotting a very simple quadratic curve $y = x^2$, with x with integer values from $0$ to $5$. We use here `numpy` arrays that are more versatile with operands.

```python
x = np.arange(0,6)
y = x**2
```

`matplot` lib `plot()` function is probably the simplest way to create a _line chart_ of points $(x,y)$ within cartesian axes. If you just pass `numpy` arrays $x$ and $y$ as parameters to this method, and then you call the `show()` method to invoke the plot you created, you are basically done.

```python
plt.plot(x,y)
plt.show()
```

If you want to improve the aesthetics of the chart, you usually call some other methods after `plot` and before `show()`. For example, `xtitle()`, `xlabel()` and `ylabel()` add respectively title, x-axis name, and y-axis name to the plot previously created. 

```python
plt.plot(x,y)
plt.title("A portion of a quadratic curve")
plt.xlabel("$X$")
plt.ylabel("$Y = X^2$")
plt.show()
```

You will learn that `matplotlib` methods are very powerful and they adapt themselves in the most convenient way without excessively tuning with parameters and other functions. You can for example add negative values to our quadratic function, and the plot will be shown approximately as you expect without changing almost anything.

```python
x1 = np.arange(-15,16)
y1 = x1**2
plt.plot(x1,y1)
plt.title("A portion of a quadratic curve")
plt.xlabel("$X$")
plt.ylabel("$Y = X^2$")
plt.show()
```

It is pretty straightforward to change the size of the figure: you need to pass the length of rows and columns to method `figure()` using argument `figsize`. Just remember that this argument wants a tuple.

```python
plt.figure(figsize=(15,5))
plt.plot(x,y)
plt.show()
```

Changing color is also very simple. There is the `fmt` optional paramater of the `plot()` method that can be used to pass format strings, to be interpreted as just an abbreviation for setting basic line properties. All of these and more can also be controlled by keyword arguments. For example, `"r-"` will plot a red line.
For a complete description of the `plot()` method, read the official [matplotlib plot() manual](https://matplotlib.org/3.2.0/api/_as_gen/matplotlib.pyplot.plot.html)

```python
plt.plot(x,y,"r-")
plt.show()
```

## Scatter plot

There is a method for creating scatter plots that is called (guess!) `scatter()`. However, if you want to stick for a little while to out `plot()` method, you can just decide to play a little bit with the second character of the `fmt` string. If you use a `o` instead of a `-`, you will plot dots of any color you like (and green is the color if you use `g`).

```python
plt.plot(x1,y1,"go")
plt.title("A scattered portion of a quadratic curve")
plt.xlabel("$X$")
plt.ylabel("$Y = X^2$")
plt.show()
```

Scatter plots are used very often to make sense of (raw) data without any help from lines, bars, histograms, and so on. When data is particularly noisy, scatter plots will return the beauty of absolute confusion. 
If you create plots from randomly generated values, you can exercise yourself in fighting your own bias (remember the gestalt effect of the pareidolia phenomenon?). For example, can you find some kind of _order_ or a _familiar shape_ in the following plot?

```python
import random
yr= np.random.randint(-100, 100, size=(31))
plt.plot(x1,yr,"b+") # using blue plusses '+' in place of circles 'o'
plt.title("A scatter plot of randomly generated numbers")
plt.xlabel("$X$")
plt.ylabel("$Y$")
plt.show()
```

Can you see some pattern in the plot above? Well, if you found it, you are probably wrong.
If you ask what happens if you use `scatter()` method instead of the more generic `plot()`, just see below.

```python
plt.scatter(x1,yr,color="blue", marker="+") # using blue plusses '+' in place of circles 'o'
plt.title("A scatter plot of randomly generated numbers")
plt.xlabel("$X$")
plt.ylabel("$Y$")
plt.show()
```

Apparently, it does not change much. We cannot use `fmt` argument to make things faster, but the semantic is clear because we have to use pre-defined arguments. For example the aesthetics can be controlled at a finer grained level. See [matplotlib scatter() manual](https://matplotlib.org/3.2.1/api/_as_gen/matplotlib.pyplot.scatter.html) for the official documentation.

Let's go back to our `plot()` method, because you can make other interesting things with it and very few lines of codes. For example, you can plot multiple sets of data by passing in multiple sets of arguments of X and Y axis.

```python
y3 = x**3
```

```python
plt.plot(x,y,"go", x, y3, "r^")
plt.title("Plotting multiple sets of data")
plt.xlabel("X")
plt.ylabel("$Y = X^2$ and $Y=X^3$")
plt.show()
```

## Plotting multiple charts

If you want to plot more than one $(x,y)$ points series, it is probably useful if you learn to use `subplot()`. You can have one separate plot for each sets of data, without sharing the same axes as we did above. You need to pass three arguments: `nrows`, `ncols` and `index`. Quite trivially, they can be used to place each plot in a cell of a bigger plot matrix, as they indicate the number of rows, number of columns and the index number of the sub-plot. 
In our example, we want to place the second plot at the right of the first. So we want to place them in one row and two columns. The cell number is indicated as the third argument. Hence we pass arguments `(1,2,1)` and `(1,2,2)` to the `sublot` methods that must be called before the given `plot()` method for each chart. You have also to call separately `title()` and other methods, using `suptitle()` method to make a centralized title for the whole figure.

```python
plt.subplot(1,2,1)
plt.plot(x,y,"go")
plt.title("A quadratic function")
plt.subplot(1,2,2)
plt.plot(x, y3, "r^")
plt.title("A cubic function")
plt.suptitle("Comparing quadratic and cubic functions")
plt.show()
```

If you want to place your subplots in different positions, just use another grid setting the `nrows`, `ncols` and `index` arguments properly

```python
plt.subplot(2,1,1)
plt.plot(x,y,"go")
plt.title("A quadratic function")
plt.subplot(2,1,2)
plt.plot(x, y3, "r^") # plotting red diamonds with "r^"
plt.title("A cubic function")
plt.suptitle("Comparing quadratic and cubic functions")
plt.show()
```

When you have to create many subplots, using the above method can be definitely a little bit boring. If you want to leave part of the work to python, then `subplots()` with the `s` is the method you are looking for. It takes two arguments `nrows` and `ncols` as number of rows and number of columns respectively, and returns two objects: a `figure` and `axes` that you can store in two separate variables. They can be used to change the figure and axes attributes. 
Please observe that when you deal with axes objects, you have to use `get_` and `set_` methods. Hence, instead of `title()` use `set_title()`.

```python
x = np.arange(1,16)

fig,ax=plt.subplots(nrows=2,ncols=2,figsize=(10,10))

y = x**2
ax[0,0].plot(x,y,"go")
ax[0,0].set_title("Squares")

y1 = x**3
ax[0,1].plot(x, y1, "r^")
ax[0,1].set_title("Cubes")

y2 = np.log(x)
ax[1,0].plot(x,y2,"b-")
ax[1,0].set_title("log")

y3 = np.log10(x)
ax[1,1].plot(x, y3, "k+")
ax[1,1].set_title("log10")

plt.suptitle("Comparing different growth")

plt.show()
```

## Encouraging comparisons

We have to admit that something went wrong. We are comparing different curves, and apparently their behavior is comparable. The problem is that we are using different scales for the y-axes, and this does not encourage comparisons. It is likely we cannot easily compare quadratic and cubic growth with logorithmic curves. But, at least, we can set the same y-axes scales to pair of charts in our matrix. We can easily do that using `get_ylim()` and `set_ylim()` methods.

```python
x = np.arange(1,16)

fig,ax=plt.subplots(nrows=2,ncols=2,figsize=(10,10))

ax[0,0].plot(x,y,"go")
ax[0,0].set_title("Squares")

ax[0,1].plot(x, y1, "r^")
ax[0,1].set_title("Cubes")

ax[0,0].set_ylim(ax[0,1].get_ylim())

ax[1,0].plot(x,y2,"b-")
ax[1,0].set_title("log")

ax[1,1].plot(x, y3, "k+")
ax[1,1].set_title("log10")
ax[1,1].set_ylim(ax[1,0].get_ylim())

plt.suptitle("Comparing different growth")

plt.show()
```

## Other simple chart

We will learn with the next notebooks how to use other methods to plot basic charts like `bar()`, `hist()`, `pie()`, and so on. Here we want to introduce *area charts*, that can be easily added to *line charts* with the `fill_between()` method. Let's see an example when plotting a classic gaussian curve. 

```python
from scipy.stats import norm

x = np.arange(-10, 10, 0.001)
y = norm.pdf(x,0,3) # Mean = 0, Std Dev = 3.

plt.plot(x, y)
plt.fill_between(x,y)
plt.show()
```

Can you do better? I bet yes. If you recall one of the Coline Ware's famous guidelines: *Use more saturated colors when color coding small symbols, thin lines, or other small areas. Use less saturated colors for coding large areas*. You have a thin line that is your curve, that actually blends with the area below. Just use differently saturated shades of blue of the line and the area and compare the results.

```python
plt.plot(x, y, color="skyblue", alpha=0.8)
plt.fill_between(x,y, color="skyblue", alpha=0.2)
plt.show()
```

Normal distributions are so beautiful and theoretically perfect. Unfortunately they are very rare when it comes to describe natural phenomena. Power laws are equally exceptional mathematical creatures, and they characterise a lot of different domains. Let's plot a power law distribution: we create a scatter plot of $(k,p(k))$ points, s.t., $p(k) = k^{-2.5}$.

```python
k = np.arange(1,1000)
f = pow(k,-2.5)
plt.plot(k,f,"r-")
plt.title("power law")
plt.xlabel("k")
plt.ylabel("$p(k) = k^-2.5$")
plt.show()
```

If you are disappointed, you are definitely right. The curve is hardly visible, and you cannot find out why everybody says that power law are so meaningful. The main problem with power law function is that they decreases so rapidly that you cannot understand that if you mantain a linear to linear scale. Fortunately enough you can swithc to a log scale for both of your axes using `xscale()` and `yscale()` methods.

```python
k = np.arange(1,1000)
f = pow(k,-2.5)
# plotting the (k,f) points with both a line chart and a scatter plot
plt.plot(k,f,"r-", k,f,"ko")
plt.title("power law in a log-log plot")
plt.xscale("log")
plt.yscale("log")
plt.xlabel("k")
plt.ylabel("$p(k) = k^-2.5$")
plt.show()
```

Don't expect that real and raw data return such perfect lines. However, now you got the point. In a log-log plot a power law is just a straight line. Can you mathematically prove it to understand why?
